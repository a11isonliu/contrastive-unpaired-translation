```python
%matplotlib ipympl
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
file_savepath = '/media/faraday/magnetograms_fd/'
ds = 'mdi2hmi_small'
trainA_path = './' + ds + '/trainA.csv'
trainB_path = './' + ds + '/trainB.csv'
mdi_ds = pd.read_csv(trainA_path, header=None)
mdi_ds['date'] = pd.to_datetime(mdi_ds[0].str.split('.').str[2].str.strip('_TAI'), format='%Y%m%d_%H%M%S')
hmi_ds = pd.read_csv(trainB_path, header=None)
hmi_ds['date'] = hmi_ds[0].str.split('.').str[2].str.strip('_TAI')
mdi_ds.head(5)
```

```python
def get_magnetogram(n):
    return np.load(file_savepath + 'mdi_fd/' + mdi_ds[0][n]), mdi_ds['date'][n]

n = 4
data, date = get_magnetogram(n)
plt.imshow(data)
print(date)
plt.show()
```
